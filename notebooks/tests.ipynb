```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.linear_model import LinearRegression, LogisticRegression

# Create a dataset for classification
X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=42)
colors = ['red' if label == 0 else 'blue' for label in y]

# Train a simple logistic regression classifier for the decision boundary
clf = LogisticRegression().fit(X, y)

# Create a grid to plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict class using data and logistic regression classifier
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create a dataset for regression
X_reg = np.random.rand(100, 1) * 10  # Random data points
y_reg = 2.5 * X_reg + np.random.randn(100, 1) * 2  # Linear relationship with noise
reg_model = LinearRegression().fit(X_reg, y_reg)
y_pred = reg_model.predict(X_reg)

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Classification plot
ax[0].contourf(xx, yy, Z, alpha=0.8)
ax[0].scatter(X[:, 0], X[:, 1], c=colors)
ax[0].set_title("Przykład klasyfikacji")
ax[0].set_xlabel("Cecha 1")
ax[0].set_ylabel("Cecha 2")
ax[0].grid()

# Regression plot
ax[1].scatter(X_reg, y_reg, color='gray')
ax[1].plot(X_reg, y_pred, color='red')
ax[1].set_title("Przykład regresji")
ax[1].set_xlabel("Cecha")
ax[1].set_ylabel("Cel")
ax[1].grid()

plt.tight_layout()
plt.show()
```
